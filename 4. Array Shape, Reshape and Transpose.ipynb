## 4. Array Shape, Reshape and Transpose

```python
# Importing numpy library
import numpy as np
```

```python
# Creating a array
cloth_arr = np.array([['Kurta & dhoti', 'Angvastra', 'Pancha'],
                      ['silk saree', 'chudi daar', 'langa voni']])

cloth_arr
```

```python
# reshaping the the cloth_arr to a 3X2 array 
# 2D array

cloth_arr.reshape(3,2)
```

```python
# reshaping the the cloth_arr to a 1X6 array
# 2D array

cloth_arr.reshape(1,6)
```

```python
# reshaping the the cloth_arr to a 6X1 array
# 2D array

cloth_arr.reshape(6,1)
```

```python
# reshaping the the cloth_arr to a 1X2X3 array
# 3D array

cloth_arr.reshape(1,2,3)
```

```python
# ravel() : Flatten numpy array to 1D array

d = cloth_arr.ravel()
d
```
