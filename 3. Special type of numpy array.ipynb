## 3. Special Type of Numpy Array

```python
# importing the numpy library
import numpy as np
```

```python
# Numpy array which is initialized with zero

# 1D zero array
zero_arr = np.zeros(5)
zero_arr
```

```python
# 2D zero array
zero_arr = np.zeros((2,3))
zero_arr
```

```python
# 3D zero array
zero_arr = np.zeros((2,3,3))
zero_arr 
```

```python
# create numpy array which is initialized with ones

# 1D ones array
ones_arr = np.ones(6)
ones_arr
```

```python
# 2D ones array
ones_arr = np.ones((2,2))
ones_arr
```

```python
# 3D ones array
ones_arr = np.ones((2,2,2))
ones_arr
```

```python
# create a numpy array with a specific value
# full() : this numpy functions helps in creating a numpy array with a specific value

# 1D array filled with 6
array = np.full(5,6)
array
```

```python
# 2D array filled with 9
array = np.full((3,3),9)
array
```

```python
# 3D array filed with 7
array = np.full((2,3,2),7)
array
```

```python
# identity matrix
# where diagonal elements of the matrix is 1 and other elements are 0

arr = np.eye(5)
arr
```

```python
# random() package in numpy used for generating random elements
```

```python
# 1D array of random elements
# rand() : generates random elements between 0 to 1
# rand() function using uniform distribution while selecting the values

np.random.rand(10)
```

```python
# generate a 2D array of random elements (0 to 1) using rand()

np.random.rand(2,3)
```

```python
# generate a 3D array of random elements (0 to 1) using rand()

np.random.rand(2,2,3)
```

```python
# random() works same as rand()
# difference between rand and random is random accepts only single positional argument or a tuple
# and rand accepts multiple arguments

# 1D array

np.random.random(3)
```

```python
# 2D array

np.random.random((2,3))
```

```python
# 3D array

np.random.random((2,3,4))
```

```python
# randn() : Generates most of the values between -3 to +3
# here n stands for standard Normal distribution (mean = 0, sd = 1)

# 1D Array

np.random.randn(8)
```

```python
# 2D Array

np.random.randn(2,3)
```

```python
# 3D Array

np.random.randn(2,2,3)
```

```python
# randint() : this will generate random integer elements between a range
# randint(start_range, end_range, array_shape)
# 1D Array

rand_int = np.random.randint(10, 100, 5)
rand_int
```

```python
# 2D Array
rand_int = np.random.randint(10, 100, (2,3))
rand_int
```

```python
# 3D Array
rand_int = np.random.randint(10, 20, (2,6,2))
rand_int
```

```python
# choice() : works on categorical data
# choice() function is used for selecting random elements from categorical list

ch = np.array(['vinnu','chinnu','binnu','berry'])
np.random.choice(ch, size = 5)
```

```python
ch = np.array([12.5, 67.9, 34.556, 90.89])
np.random.choice(ch, 6)
```

```python
### ???????
ch = np.array([1,2,3,4,5])
np.random.choice(ch, size = 4)
```

```python
# here replace default value is True means meaning that a value of the list can be selected multiple times.
# replace = False means vice versa of replace = True
np.random.choice(ch, size = 4, replace = False)
```

```python
np.random.choice(ch, size = 10, p = [0.1, 0.3, 0.4, 0.15, 0.05])
```

```python
# generates the random values starting from the seed value
# and generated random values are always same

np.random.seed(42)
np.random.choice(ch, size = 4)
```

```python
# Create Sequence, Shuffle, Permutation, Linearly Spaced
```

```python
# arrange() : Create sequence of elements

arr1 = np.arange(10,1,-1) # arange(start, end, step)
arr1
```

```python
arr1 = np.arange(1,10,1)
arr1
```

```python
# shuffle() : used for shuffling data inplace
# any new array is not created the same array is shuffled that is called inplace
# shuffle function rearrange the elements in the original array

np.random.shuffle(arr1)
arr1
```

```python
# permutation() : re_arrange the elements
# permutation() function return a new shuffled array

arr2 = np.random.permutation(arr1)
arr2
```

```python
# linspace(): linearly spaced numpy array
# this generates a special type of numpy array where every two elements have same difference

arr3 = np.linspace(1, 10, 5)
arr3
```
