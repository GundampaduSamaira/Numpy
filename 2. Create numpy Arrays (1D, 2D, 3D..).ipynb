## 2. Create numpy Arrays (1D, 2D, 3D..)

```python
# importing numpy library
import numpy as np
```

```python
# 1D Numpy Array

arr1 = np.array([1,2,3,45,67])
print(arr1)
print(type(arr1))
print(arr1.dtype)
```

```python
# how to check the dimension of array
print(arr1.ndim)
```

```python
# if different datatypes are given then while converting the given list to 
# numpy array numpy tries to change all the elements in the list to same 
# datatype

arr1 = np.array([1,2,3,4,5,6.7])
print(arr1)
print(type(arr1))
```

```python
# to check the datatype of numpy array
print(arr1.dtype)
```

```python
arr1 = np.array([1,2,3,4,'hello',5.5])
print(arr1)
print(type(arr1))
print(arr1.dtype) # U32 : unicode string type with 32 length
```

```python
# 2D Array

arr2 = np.array([['A','B','C'],['D','E','F']])
arr2
```

```python
print(arr2.ndim)
```

```python
arr2 = np.array([[1,2,3],[4,5,6]])
arr2
```

```python
arr2.ndim
```

```python
# to find the rows and columns of the array
arr2.shape
```
