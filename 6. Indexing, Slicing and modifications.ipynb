## 6. Indexing, Slicing and modifications

```python
# importing numpy library
import numpy as np
```

```python
sweets = np.array([["kaju katli", "mysore paak", "Khaja"],["bandar laddo", "Rasogula", "Ghevar"]])
sweets
```

```python
print(sweets[1,2])
```

```python
print(sweets[0,:])
```

```python
print(sweets[:,1])
```

```python
print(sweets[-1][-1])
```

```python
print(sweets[-1])
```

```python
print(sweets[::-1])
```

```python
# Modification / Update

sweets[1,0] = "payasam"
sweets
```
