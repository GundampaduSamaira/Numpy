## 5. Size and Datatype Conversion

```python
# importing numpy library
import numpy as np
```

```python
songs = np.array(["Monicaaaaaaaaaaaa", "Hukum", "Devara"])
songs
```

```python
songs.size
```

```python
# U : unicode string of max length 6
print(songs.dtype)
```

```python
# astype() : Data type conversion
song_duration = np.array([4.223, 5.4, 100.4])
int_arr = song_duration.astype(int)
int_arr
```

```python
song_earning = np.array(["500", "300", "5000"])
song_earning.dtype
```

```python
# string to integer conversion
r1 = song_earning.astype(int)
r1
```

```python
r1.dtype
```

```python
r2 = song_earning.astype(float)
r2
```

```python
r2.dtype
```

```python
np.array([True, False, True]).astype(int)
```

```python
np.array([True, False, True]).astype(float)
```

```python
np.array([1,1,0,1,0,0]).astype(bool)
```

```python
np.array([12,6,80,-4,0,56,0]).astype(bool)
```
